### Villkorssatser
 
En villkorssats innebär att ifall något villkor uppfylls händer något. 
 
Till exempel "Ifall jag är hungrig äter jag" är en villkorssats.

Villkorssatser i programmering utförs med hjälp av jämförelseoperatorer. Pröva att köra exemplet nedan och ge programmet olika input.


```
#Vi ber användaren om två personers åldrar och sparar dem i variablerna alder_1 och alder_2:
alder_1 = int(input("Hur gammal är den första personen?"))
alder_2 = int(input("Hur gammal är den andra personen?"))

#Med hjälp av en villkorssats undersöker vi ifall den första personen är äldre än den andra.
if alder_1 > alder_2:
    #Ifall villkoret uppfylls skriver programmet ut "Person ett är äldre än person två"
    print("Person ett är äldre än person två")
```


I Python används kommandot if för att skapa villkorrssatser. Lägg märke till nästa radens indrag. Indraget berättar vilka rader som utförs ifall villkoret uppfylls. Alla de rader som är indragna direkt efter if-satsen utförs.

Förutom if-kommandot kan man även använda elif och else-kommandon. Elif är en förkortning av orden else if (annars ifall).

Till exempel "Ifall jag är hungrig äter jag. Annars ifall jag är törstig dricker jag. Annars tittar jag på television"

Pröva att köra exemplet nedan genom att ge olika input till programmet.

```
#vi ber om användarens ålder och sparar den i variabeln alder:
alder=int(input("Hur gammal är du?"))

#Med hjälp av en villkorssats bestämmer vi ifall användaren är under 16 år:
if alder<16:
    #Ifall villkoret uppfylls skriver vi ut "Du är under 16 år"
    print("Du är under 16 år")
# Ifall användaren var över 16 år undersöker vi ifall hen ändå är minderårig:
elif alder<18:
    print("Du är redan över 16 år men ändå minderårig")
#Ifall användaren inte var minderårig går vi till else-satsen:
else:
    print("Du är myndig!")
```

For-loopen lönar sig att använda då man vet hur många gånger man vill upprepa en bit kod. For-loopen används även allmänt för att gå igenom olika listor.

Kör exemplen nedan och pröva på att byta talet i range-funktionen. Vad händer?

### 3.2 While-loopen

While-loopen är en iterationssats som upprepar en bit kod tills något villkor uppfylls. While-loopar används  alltså tillsammans med jämförelseoperatorer.

```
#Vi vill igen hälsa på användaren flera gånger, denna gång sammanlagt 5 gånger och med hjälp av en while-loop

#vi gör en variabel som innehåller information om hur många gånger det hälsats på användaren, alltså hur
#många gånger koden är upprepad
#I början har koden inte upprepats och därmed ställer vi variabelns värde till noll

ganger = 0

#Vi ville hälsa 5 gånger alltså skall loopen upprepas så länge ganger<5 :
while ganger<5:
  #i loopen skriver vi först ut den önskade hälsningen
  print("Hej!")
  #nu lägger vi till 1 till ganger-variabeln för vi har hälsat på användaren en gång till
  ganger = ganger + 1
```

I exemplet ovan visste vi upprepningarnas antal men i allmänhet lönar det sig att använda en while-loop då man inte vet upprepningarnas mängd på förhand. 


```
#Du har 4€ och du vill köpa så många 30cents slickepinnar som möjligt.
#Du kan inte beräkna hur många slickepinnar du får med hjälp av huvudräkning så du skriver ett program
#som beräknar hur många slickepinnar du kan köpa med hjälp av en while-loop

#vi skapar först en variabel pengar i vilken vi sparar informationen om hur mycket pengar vi har kvar
#i början har vi 4€
pengar = 4

#vi skapar även en variabel slickepinnar i vilken vi sparar mängden köpta slickepinnar
#i början har vi 0 slickepinnar
slickepinnar = 0

#vi köper slickepinnar så länge vi har råd alltså så länge som vi har minst 30 cent:
while pengar>=0.3:
  #om det finns tillräckligt med pengar köper vi en slickepinne till
  slickepinnar = slickepinnar + 1
  #eftersom du köpte en slickepinne subtraherar vi slickepinnens pris från dina pengar
  pengar = pengar - 0.3

#till sist skriver vi ut hur många slickepinnar du köpte:
print(slickepinnar, "slickepinnar")
#vi skriver även ut hur mycket pengar som blev över:
print( pengar, "euro blev över.")
```
